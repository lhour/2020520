```python
##  基础函数库
import numpy as np 

## 导入画图库
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
pd.read_csv('train.csv')
```

```python
traindata = pd.read_csv('train.csv')
```

```python
df = pd.DataFrame(traindata)
```

```python
df
```

```python
#将所有特征拿出来进行处理
df['heartbeat_signals']
```

```python
#分割，转变成数字类型
traindata_f = []
for i in df['heartbeat_signals']:
    j = i.split(',')
    traindata_f.append(j)
```

```python
#查看第一行各项特征
traindata_f[0]
```

```python
#数据处理
traindata_s = np.array( traindata_f , dtype=np.double)
train_class = np.array(df['label'])
```

```python
traindata_s[0]
```

```python
plt.figure()
fig, axs = plt.subplots(4,10,sharex=True,sharey=True,figsize=(40,10))
x_data = np.arange(len(traindata_f[0]))
al = []
lei = {0.0:'the first',1.0:'the sec',2.0:'the thr',3.0:'the four'}
for i in range(4):
    for j in range(10):
        for k in range(len(train_class)):
            if train_class[k] == i and k not in al:
                axs[i][j].plot(x_data,traindata_s[k])
                if j == 0:
                    axs[i][j].set_ylabel(lei[train_class[k]])
                al.append(k)
                break



plt.show()
```

```python
from sklearn.model_selection import train_test_split
## 从sklearn中导入逻辑回归模型
from sklearn.linear_model import LogisticRegression
## 定义 逻辑回归模型
clf = LogisticRegression(random_state=0, solver='lbfgs')

x_train, x_test, y_train, y_test = train_test_split(traindata_s,train_class, test_size = 0.2, random_state = 2020)
```

```python
clf.fit(x_train,y_train)
```

```python
## 查看其对应的w
print('the weight of Logistic Regression:',clf.coef_)

## 查看其对应的w0
print('the intercept(w0) of Logistic Regression:',clf.intercept_)
```

```python
from sklearn import metrics
train_predict = clf.predict(x_test)
print('The accuracy of the Logistic Regression is:',metrics.accuracy_score(y_test,train_predict))
```

```python
from sklearn import metrics
train_predict = clf.predict(x_train)
print('The accuracy of the Logistic Regression is:',metrics.accuracy_score(y_train,train_predict))
```

```python
#列出各个可能性
train_predict_proba = clf.predict_proba(x_train)
print('The test predict Probability of each class:\n',train_predict_proba)
```

```python
test_csv = pd.read_csv('testA.csv')
df2 = pd.DataFrame(test_csv)
testdata = []
for i in df2['heartbeat_signals']:
    j = i.split(',')
    testdata.append(j)
testdata_F = np.array(testdata,dtype=np.double)
# testdata_F[0]
```
